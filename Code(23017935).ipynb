Handling missing values and converting them to proper format

```python
# Fill or drop missing values (example)
df.fillna(df.median(numeric_only=True), inplace=True)  # Fill numeric missing values with median
df.dropna(subset=['customer_type'], inplace=True)  # Drop rows with missing 'customer_type'
```

```python

# Identify categorical columns
categorical_cols = ['hotel', 'customer_type', 'market_segment', 'deposit_type']

# Apply Label Encoding
encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

# Verify encoding
print(df.head())
```
